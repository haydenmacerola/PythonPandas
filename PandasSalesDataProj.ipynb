```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

### Clean Data

### Remove NaN Values

```python
nan_df = all_data[all_data.isna().any(axis=1)]

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()

```

### convert quantity and price to int for multiplication

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) 
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

## Question 1 best month for sales, how much was earned

### add month column 

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Add Sale per order column

```python
all_data["Sale"] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
all_data.groupby('Month').sum()['Sale']
```

#### What City had the highest number of sales

```python
all_data['State'] = all_data['Purchase Address'].str[-8:-6]
all_data.head()
```

```python
### use.apply() to create city column, allows function across a column
### lets now drop state and edit our apply fucntion to include state and city due to dupes 
```

```python
all_data = all_data.drop(columns ='State')
```

```python
###all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split (',')[1]) original func
### updated func below

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
all_data.groupby('City').sum()['Sale']
```

### what products are most often sold together

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['GroupedOrder'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
### joins cell values of prodcut for same orderID,.transform allows use of fucntion
```

```python
df = df[['Order ID', 'GroupedOrder']].drop_duplicates()

```

```python
### stack reference: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations 
from collections import Counter

count = Counter()

for row in df['GroupedOrder']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key, value in count.most_common(10):
    print(key, value)
```
